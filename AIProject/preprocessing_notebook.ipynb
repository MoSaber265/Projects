# Preprocessing Notebook
This notebook performs preprocessing on student_exam_data.csv

```
import pandas as pd
import numpy as np

df = pd.read_csv('/mnt/data/student_exam_data.csv')
df.head()
```

## Check missing values

```
df.isnull().sum()
```

## Fill missing values

```
df['StudyHours'] = df['StudyHours'].fillna(df['StudyHours'].mean())
df['PreviousScore'] = df['PreviousScore'].fillna(df['PreviousScore'].median())
df['Extracurricular'] = df['Extracurricular'].fillna(df['Extracurricular'].mode()[0])
df['Gender'] = df['Gender'].fillna('Unknown')
df.isnull().sum()
```

## Encoding Categorical

```
df_enc = pd.get_dummies(df, columns=['Gender','Extracurricular'], drop_first=True)
df_enc.head()
```

## Scaling

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = df_enc.copy()
df_scaled[['StudyHours','PreviousScore','ExamScore']] = scaler.fit_transform(
    df_scaled[['StudyHours','PreviousScore','ExamScore']]
)
df_scaled.head()
```
